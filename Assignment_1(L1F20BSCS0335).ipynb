```python
# importing the pandas library to read data in data frame
import pandas as pd
df=pd.read_csv('jsrt_metadata.csv')
import numpy as np
df
```

```python
# checking if the file has missing values.
df.isnull().sum()
```

```python
# replacing the missing values of quatitative data with mean.
mean=df['subtlety'].astype('float').mean()
mean=int(mean)
mean
```

```python
df['subtlety'].replace(np.nan,mean,inplace=True)
```

```python
df['state'].replace('non-nodule',np.nan,inplace=True)
minimum=df['state'].value_counts().idxmin()
df['state'].replace(np.nan,minimum,inplace=True)
df
```

```python
# replacing the missing values of quatitative data with mean and qualitative data with minimum or maximum repeated data value.
mean=df['size'].astype('float').mean()
mean=int(mean)
df['size'].replace(np.nan,mean,inplace=True)
mean=df['x'].astype('float').mean()
mean=int(mean)
df['y'].replace(np.nan,mean,inplace=True)
mean=df['y'].astype('float').mean()
mean=int(mean)
df['x'].replace(np.nan,mean,inplace=True)
maximum=df['position'].value_counts().idxmax()
maximum
df['position'].replace(np.nan,maximum,inplace=True)
minimum=df['diagnosis'].value_counts().idxmin()
minimum
df['diagnosis'].replace(np.nan,minimum,inplace=True)
df.isnull().sum()
```

# QUESTION 2.


```python
# number of pateints of each disease with respect to state.
# making new data frame by joining two columns.
df['diagnosis-state']= df['state']+' '+df['diagnosis']
df['diagnosis-state'].value_counts()
```

```python
# storing labels of newly made data frame into a list.
mylabels = df['diagnosis-state'].value_counts().index.tolist()
mylabels
```

```python
# pie chart of number of pateints of each disease with respect to state
plt.pyplot.pie(df['diagnosis-state'].value_counts(),labels=mylabels,autopct='%1.1f%%',radius=5)
```
